# Set Operations in Python

This notebook demonstrates various set operations in Python:  
- **union()**  
- **intersection()**  
- **difference()**  
- **symmetric_difference()**

---

##  **Union of Two Sets**

The `union()` method combines all elements from two sets, eliminating duplicates.

### Task:
Create two sets and use the `union()` method to combine them.

**Data:**
```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
```

**Solution:**
```python
union_set = set1.union(set2)
print(union_set)
```

---

##  **Intersection of Two Sets**

The `intersection()` method returns a set containing all elements that are common to both sets.

### Task:
Use the `intersection()` method to find the common elements between two sets.

**Data:**
```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
```

**Solution:**
```python
intersection_set = set1.intersection(set2)
print(intersection_set)
```

---

##  **Difference of Two Sets**

The `difference()` method returns a set containing all elements from the first set that are not in the second set.

### Task:
Use the `difference()` method to find elements in the first set but not in the second.

**Data:**
```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
```

**Solution:**
```python
difference_set = set1.difference(set2)
print(difference_set)
```

---

##  **Symmetric Difference of Two Sets**

The `symmetric_difference()` method returns a set containing all elements from both sets except the common elements.

### Task:
Use the `symmetric_difference()` method to find elements that are in either set but not in both.

**Data:**
```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
```

**Solution:**
```python
symmetric_diff_set = set1.symmetric_difference(set2)
print(symmetric_diff_set)
```

---

##  **Conceptual Questions**

### 1. **What does the `union()` method do in Python sets?**

**Answer:**  
The `union()` method combines all the elements from two sets, eliminating any duplicates. It returns a new set with all unique elements from both sets.

---

### 2. **What is the difference between `intersection()` and `difference()` methods in Python sets?**

**Answer:**  
- **`intersection()`** returns only the elements that are common to both sets.
- **`difference()`** returns only the elements that are present in the first set but not in the second.

---

### 3. **When would you use `symmetric_difference()` instead of `union()`?**

**Answer:**  
You would use `symmetric_difference()` when you want to find elements that are in either of the sets but not in both, excluding the common elements. In contrast, `union()` would include the common elements as well.
